# Exploring eBay Car Sales Data

This project explores a dataset of used car listings scraped from **eBay Kleinanzeigen**, the classifieds section of the German eBay website. The dataset contains **50,000 listings** and includes details such as vehicle brand, model, price, mileage, fuel type, and registration year.

The primary goals of this project are to **clean the raw dataset** â dealing with messy string formatting, missing values, and outliers â and then perform **exploratory analysis** to uncover patterns in the used car market. For example, we'll investigate which brands and models are most common, how price relates to mileage and age, and whether unrepaired damage significantly affects listing prices.

The dataset contains 20 columns covering both vehicle attributes (e.g. `brand`, `model`, `powerPS`, `odometer`) and listing metadata (e.g. `dateCrawled`, `dateCreated`, `lastSeenOnline`). Some columns require cleaning before analysis, as numeric values like `price` and `odometer` are stored as formatted strings.

From the output of `info()` and `head()`, we can note the following:

## Column Renaming Summary

We renamed all 20 columns using a mapping dictionary and `autos.columns.map()`. Two types of changes were made:

**1. camelCase â snake_case** (Python convention): Column names like `dateCrawled`, `offerType`, and `fuelType` were converted to `date_crawled`, `offer_type`, and `fuel_type`. This makes them consistent with Python's style guide (PEP 8) and easier to type without accidentally triggering case-sensitive bugs.

**2. Descriptive renames**: Four columns were also given clearer names:
- `yearOfRegistration` â `registration_year` â groups both registration columns under a common `registration_` prefix
- `monthOfRegistration` â `registration_month` â same rationale
- `notRepairedDamage` â `unrepaired_damage` â more natural phrasing; the double negative in the original name was easy to misread
- `dateCreated` â `ad_created` â disambiguates it from other date columns; makes clear this is when the *listing* was created, not the car

## Descriptive Statistics â Observations

**Columns that are candidates to be dropped:**
- `n_of_pictures` â mean, min, and max are all `0`. Every single listing has zero pictures, making this column completely uninformative.
- `seller` â 49,999 out of 50,000 listings are `privat` (private). Only 1 is a dealer. No analytical value.
- `offer_type` â 49,999 out of 50,000 are `Angebot` (offer). Same issue â no variation.

**Columns needing further investigation:**
- `registration_year` â min is `1000` and max is `9999`, which are clearly erroneous. We'll need to filter to a realistic range.
- `power_ps` â max is `17,700` PS and min is `0`. Both extremes are likely data entry errors.
- `registration_month` â min is `0`, which is not a valid month (months run 1â12).
- `price` â top value is `$0` with a frequency of 1,421. Many listings may have been posted with no price. Needs cleaning.
- `odometer` â only 13 unique values, stored as strings with `km` suffix (e.g. `150,000km`). Needs to be converted to numeric.

**Numeric data stored as text:**
- `price` â stored as a string with a `$` prefix and comma separators (e.g. `$5,000`). Needs the `$` and `,` stripped and conversion to `int`.
- `odometer` â stored as a string with a `km` suffix and comma separators (e.g. `150,000km`). Needs the `km` and `,` stripped and conversion to `int`. The column should then be renamed to `odometer_km` to reflect the unit.

```python
print(autos['price'])
print(autos['odometer'])
```

```python
autos['price'] = autos['price'].str.replace('$', '')
autos['price'] = autos['price'].str.replace(',', '')

autos['odometer'] = autos['odometer'].str.replace('km', '')
autos['odometer'] = autos['odometer'].str.replace(',', '')


autos['price'] = autos['price'].astype(float)
autos['odometer'] = autos['odometer'].astype(float)
```

```python
autos.rename(columns={'odometer': 'odometer_km'}, inplace=True)
```

## Filtering `registration_year`

**Chosen bounds: 1900â2016**

- **Upper bound (2016):** Since the data was scraped in 2016, no car could legally have a registration year beyond that. There were **1,772 rows** with years after 2016 â a significant and clearly erroneous group that must be removed.
- **Lower bound (1900):** Inspecting values before 1930 revealed only **6 rows**, with years like 1910, 1927, and 1929. While technically possible (vintage cars do exist), these are so rare they're likely data entry errors. Setting the lower bound at 1900 excludes the genuinely impossible values (e.g. year 1000) while retaining any legitimate historic vehicles.

## Distribution of `registration_year` After Filtering

The cleaned distribution shows a clear and expected pattern â listings are heavily concentrated in the **1995â2010** range, peaking around **1999â2005**. Pre-1980 vehicles make up a negligible fraction (under 0.1% combined), confirming they were outliers. The data now represents a realistic snapshot of the used car market: mostly vehicles that are 5â20 years old at the time of listing.

```python
brand_counts = autos['brand'].value_counts(normalize=True)
top_brands = brand_counts[brand_counts >= 0.05].index
```

```python
brand_dict = {}

for b in top_brands:
    brand_dict[b] = autos[autos['brand'] == b]['price'].mean()
        

pd.Series(brand_dict).sort_values(ascending=False)
```

```python
brand_table = pd.Series(brand_dict).to_frame('mean_price')
```

## Brand Analysis â Mean Price and Listing Count

We filtered to the 6 brands that each represent more than 1% of total listings, covering the dominant players in the German used car market.

**Volume:** Volkswagen dominates with 9,180 listings â more than double the next highest brand (BMW at 4,983). This aligns with VW's status as Germany's most popular domestic brand. Audi and Ford have the fewest listings of the group at ~3,900 and ~2,800 respectively.

**Price:** There's a clear split between budget and premium brands:
- **Audi** commands the highest mean price at **~â¬9,614**, followed closely by **Mercedes-Benz (~â¬8,767)** and **BMW (~â¬8,582)** â all three are premium German manufacturers
- **Volkswagen** sits in the middle at **~â¬5,784**, reflecting its broad range from entry-level to higher-end models
- **Opel (~â¬3,394)** and **Ford (~â¬4,292)** are the most affordable on average, consistent with their positioning as mass-market budget brands

The data suggests a strong correlation between brand prestige and average listing price, even in the used car market.

## Mean Price vs. Mean Mileage by Brand

Adding average mileage alongside mean price reveals an interesting pattern â mileage is remarkably consistent across all six brands, ranging from ~123,000km (Ford) to ~133,000km (BMW), a spread of only ~10,000km. Despite this, there are dramatic differences in mean price.

This tells us that **mileage alone does not explain the price gap between brands**. BMW and Mercedes-Benz listings actually have the *highest* average mileage yet still command prices more than double those of Opel and Ford. The price differences are better explained by **brand prestige and perceived quality** rather than how worn-out the cars are.

Opel stands out as the cheapest brand on average (~â¬3,394) despite having similar mileage to the premium brands â suggesting buyers heavily discount Opel regardless of condition. Audi leads on price (~â¬9,614) with middle-of-the-pack mileage, reinforcing its premium positioning in the used car market.
